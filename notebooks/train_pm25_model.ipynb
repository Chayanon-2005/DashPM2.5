```python
import pandas as pd
from pycaret.regression import setup, compare_models, save_model
```

```python
# โหลดข้อมูลจากไฟล์ PM2.5.csv
data_file = "../data/PM2.5.csv"
df = pd.read_csv(data_file)
```

```python
# แปลงคอลัมน์ timestamp เป็น datetime
df["timestamp"] = pd.to_datetime(df["timestamp"])
```

```python
# ตั้งค่า Pycaret และฝึกโมเดล
exp = setup(df, target="pm_2_5", session_id=123, verbose=False)
best_model = compare_models()
```

```python
# บันทึกโมเดลเป็นไฟล์ .pkl
save_model(best_model, "../models/pm25_forecast_model.pkl")

print("โมเดลถูกบันทึกเรียบร้อยแล้ว!")
```
